## Plotting

Let's define the following function $f(\mathbf{x}) = f(x_1, x_2) = (x_1 - 2)^2 + (x_2 - 1)^2$

```python
import numpy as np

f = lambda x: (x[...,0] - 2)**2 + (x[...,1] - 1)**2 # this is our function we want to plota
```

```python
# simple 2D example
x1 = np.linspace(-5,5,100)
x2 = np.linspace(-5,5,100)
x1v,x2v = np.meshgrid(x1, x2)
x = np.column_stack((x1v.reshape(-1,1), x2v.reshape(-1,1)))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=[12,8])
c=plt.contour(x1v, x2v, f(x).reshape(100,100), levels=25, colors='k') # 'k' means black color
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
# plt.colorbar(c)
plt.clabel(c, c.levels, inline=True, fontsize=14)
plt.tight_layout()
```

## Norms

```python
x = np.array([1.0, 1.0])

print(f"L2 norm of x {np.linalg.norm(x, 2)}")
print(f"Linf norm of x {np.linalg.norm(x, np.inf)}")
print(f"L1 norm of x {np.linalg.norm(x, 1)}")
```

## Derivatives

```python
import jax.numpy as jnp # this is a thin wrapper to NumPy within JAX
from jax import grad, hessian
```

```python
# computingBaseException gradient
f_ = lambda x: f(x)
gradf = grad(f_)  # Obtain its gradient function
print(f'gradient at x1: {gradf(x)}')
print(f'normalized gradient at x1: {gradf(x) / np.linalg.norm(gradf(x))}')
```

```python
# computing Hessian
hess = hessian(f)#
print(f'Hessian at x1: {hess(x)}')
```

```python
# Convert to NumPy arrays
B = hess(x)
B_np = np.asarray(hess(x))
print(f"Type of B {type(B)}") # JAX Array type
print(f"Type of B_np {type(B_np)}") # NumPy array type. Do this if the rest of your code uses NumPy.
```

### Alternative approach--using PyTorch

```python
import torch
from torch.autograd.functional import jacobian, hessian
```

```python
x_tensor = torch.tensor(x, dtype=torch.float32)
# gradient
g = jacobian(f, x_tensor)
print(f'Gradient {g}')

H = hessian(f, x_tensor)
print(f'Hessian {H}')
```

```python
# Convert to NumPy array
g_np = g.numpy()
H_np = H.numpy()
print(f"Type of g {type(g)}")
print(f"Type of g_np {type(g_np)}")
```
