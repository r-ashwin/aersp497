# Nelder-Mead on the Branin-Hoo function

This notebook is a geometric, step-by-step illustration of the **Nelder-Mead simplex method** on the two-dimensional **Branin-Hoo** test function.

The focus is on:

1. how the simplex is formed,
2. how the worst vertex is identified,
3. how reflection / expansion / contraction / shrink are proposed,
4. how the next simplex is chosen,
5. how this relates to SciPy's `Nelder-Mead` implementation.

The notebook uses the Branin-Hoo function because it is low-dimensional, smooth, and easy to visualize with contours.

## 1. Imports and plotting defaults

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.optimize import minimize

plt.rcParams.update({
    "figure.figsize": (8, 6),
    "axes.grid": True,
    "grid.alpha": 0.25,
    "font.size": 12,
    "text.usetex": True,
})

np.set_printoptions(precision=4, suppress=True)
```

## 2. The Branin-Hoo function

We use the standard form

$
f(x_1,x_2)
=
a\left(x_2-bx_1^2+cx_1-r\right)^2 + s(1-t)\cos(x_1) + s,
$

with the classical constants
$
a=1,\quad
b=\frac{5.1}{4\pi^2},\quad
c=\frac{5}{\pi},\quad
r=6,\quad
s=10,\quad
t=\frac{1}{8\pi}.
$

A common plotting box is
$
x_1\in[-5,10],\qquad x_2\in[0,15].
$

```python
def branin(x):
    x1, x2 = np.asarray(x)
    a = 1.0
    b = 5.1 / (4.0 * np.pi**2)
    c = 5.0 / np.pi
    r = 6.0
    s = 10.0
    t = 1.0 / (8.0 * np.pi)
    return a * (x2 - b * x1**2 + c * x1 - r)**2 + s * (1 - t) * np.cos(x1) + s

x1 = np.linspace(-5, 10, 400)
x2 = np.linspace(0, 15, 400)
X1, X2 = np.meshgrid(x1, x2)
Z = np.vectorize(lambda a, b: branin([a, b]))(X1, X2)

plt.figure(figsize=(8, 6))
cs = plt.contour(X1, X2, Z, levels=40)
plt.clabel(cs, inline=True, fontsize=8)
plt.xlabel(r"$x_1$")
plt.ylabel(r"$x_2$")
plt.title("Branin-Hoo contour plot")
plt.show()
```

## 3. A reminder: what Nelder-Mead does

In two dimensions, the method maintains a **triangle** (a 2-simplex) with three vertices.

At each iteration it:

- orders the vertices by objective value,
- labels them as best, second-worst, and worst,
- computes the centroid of all vertices except the worst one,
- tries a **reflection** point,
- and, depending on how good that point is, may instead accept
  - an **expansion** point,
  - an **outside contraction** point,
  - an **inside contraction** point,
  - or perform a **shrink** step.

SciPy's classical parameter values are

$
\rho=1,\qquad \chi=2,\qquad \psi=\tfrac12,\qquad \sigma=\tfrac12,
$

corresponding to reflection, expansion, contraction, and shrink.

## 4. A pedagogical trace of the SciPy-style Nelder-Mead steps

SciPy's public interface (`scipy.optimize.minimize(..., method="Nelder-Mead")`) returns the final optimization result, but it does not directly expose all intermediate trial points needed for geometric visualization.

So below, we implement that follows the same standard Nelder-Mead logic and parameter choices used by SciPy. This makes it possible to inspect:

- the simplex at the beginning of each iteration,
- the centroid,
- the reflection / expansion / contraction candidates,
- the accepted move,
- and the updated simplex.

```python
def nelder_mead_trace(
    fun,
    x0,
    initial_simplex=None,
    maxiter=20,
    xatol=1e-8,
    fatol=1e-8,
    alpha=1.0,
    gamma=2.0,
    rho=0.5,
    sigma=0.5,
):
    x0 = np.asarray(x0, dtype=float)
    n = len(x0)

    if initial_simplex is None:
        nonzdelt = 0.05
        zdelt = 0.00025
        sim = np.empty((n + 1, n), dtype=float)
        sim[0] = x0
        for k in range(n):
            y = np.array(x0, copy=True)
            if y[k] != 0:
                y[k] = (1 + nonzdelt) * y[k]
            else:
                y[k] = zdelt
            sim[k + 1] = y
    else:
        sim = np.asarray(initial_simplex, dtype=float).copy()
        if sim.shape != (n + 1, n):
            raise ValueError("initial_simplex must have shape (n+1, n)")

    fsim = np.array([fun(x) for x in sim])
    history = []

    for it in range(maxiter):
        ind = np.argsort(fsim)
        sim = sim[ind]
        fsim = fsim[ind]

        best = sim[0].copy()
        worst = sim[-1].copy()
        second_worst = sim[-2].copy()
        centroid = np.mean(sim[:-1], axis=0)

        step = {
            "iter": it,
            "simplex_before": sim.copy(),
            "f_before": fsim.copy(),
            "best": best.copy(),
            "second_worst": second_worst.copy(),
            "worst": worst.copy(),
            "centroid": centroid.copy(),
            "action": None,
            "trial_points": {},
        }

        if np.max(np.abs(sim[1:] - sim[0])) <= xatol and np.max(np.abs(fsim[0] - fsim[1:])) <= fatol:
            step["action"] = "terminate"
            step["simplex_after"] = sim.copy()
            step["f_after"] = fsim.copy()
            history.append(step)
            break

        xr = centroid + alpha * (centroid - worst)
        fxr = fun(xr)
        step["trial_points"]["reflection"] = (xr.copy(), fxr)

        if fxr < fsim[0]:
            xe = centroid + gamma * (xr - centroid)
            fxe = fun(xe)
            step["trial_points"]["expansion"] = (xe.copy(), fxe)

            if fxe < fxr:
                sim[-1] = xe
                fsim[-1] = fxe
                step["action"] = "expand"
            else:
                sim[-1] = xr
                fsim[-1] = fxr
                step["action"] = "reflect"

        elif fxr < fsim[-2]:
            sim[-1] = xr
            fsim[-1] = fxr
            step["action"] = "reflect"

        else:
            if fxr < fsim[-1]:
                xoc = centroid + rho * (xr - centroid)
                fxoc = fun(xoc)
                step["trial_points"]["outside_contraction"] = (xoc.copy(), fxoc)

                if fxoc <= fxr:
                    sim[-1] = xoc
                    fsim[-1] = fxoc
                    step["action"] = "outside_contract"
                else:
                    step["action"] = "shrink"
                    for j in range(1, n + 1):
                        sim[j] = sim[0] + sigma * (sim[j] - sim[0])
                        fsim[j] = fun(sim[j])

            else:
                xic = centroid - rho * (centroid - worst)
                fxic = fun(xic)
                step["trial_points"]["inside_contraction"] = (xic.copy(), fxic)

                if fxic < fsim[-1]:
                    sim[-1] = xic
                    fsim[-1] = fxic
                    step["action"] = "inside_contract"
                else:
                    step["action"] = "shrink"
                    for j in range(1, n + 1):
                        sim[j] = sim[0] + sigma * (sim[j] - sim[0])
                        fsim[j] = fun(sim[j])

        ind = np.argsort(fsim)
        sim = sim[ind]
        fsim = fsim[ind]

        step["simplex_after"] = sim.copy()
        step["f_after"] = fsim.copy()
        history.append(step)

    return {
        "x": sim[0],
        "fun": fsim[0],
        "nit": len(history),
        "history": history,
        "final_simplex": sim.copy(),
        "final_values": fsim.copy(),
    }
```

## 5. Choose an initial point and the initial simplex

```python
x0 = np.array([6.0, 14.0])

initial_simplex = np.array([
    [6.0, 14.0],
    [7.2, 14.0],
    [6.0, 12.6],
], dtype=float)

trace = nelder_mead_trace(branin, x0=x0, initial_simplex=initial_simplex, maxiter=12)

print("Best point after traced iterations:", trace["x"])
print("Best function value:", trace["fun"])
print("Number of traced iterations:", trace["nit"])
```

## 6. Helper routines for plotting simplices and trial points

```python
def draw_simplex(ax, simplex, color="k", lw=2, label=None, alpha=1.0, zorder=3):
    cyc = np.vstack([simplex, simplex[0]])
    ax.plot(cyc[:, 0], cyc[:, 1], "-o", color=color, lw=lw, alpha=alpha, label=label, zorder=zorder)

def contour_background(ax):
    ax.contour(X1, X2, Z, levels=30, linewidths=0.8)
    ax.set_xlim(-5, 10)
    ax.set_ylim(0, 15)
    ax.set_xlabel(r"$x_1$")
    ax.set_ylabel(r"$x_2$")

def plot_nm_step(step, annotate=True):
    fig, ax = plt.subplots(figsize=(8, 6))
    contour_background(ax)

    simp_before = step["simplex_before"]
    simp_after = step["simplex_after"]

    draw_simplex(ax, simp_before, color="tab:blue", lw=2.5, label="simplex before")
    draw_simplex(ax, simp_after, color="tab:green", lw=2.0, label="simplex after", alpha=0.9)

    best = step["best"]
    second_worst = step["second_worst"]
    worst = step["worst"]
    centroid = step["centroid"]

    ax.scatter(*best, color="tab:green", s=100, marker="o", zorder=5, label="best")
    ax.scatter(*second_worst, color="tab:orange", s=100, marker="o", zorder=5, label="second worst")
    ax.scatter(*worst, color="tab:red", s=120, marker="x", zorder=6, label="worst")
    ax.scatter(*centroid, color="k", s=80, marker="*", zorder=6, label="centroid")

    tp = step["trial_points"]
    if "reflection" in tp:
        xr, fxr = tp["reflection"]
        ax.scatter(*xr, color="purple", s=90, marker="D", zorder=6, label="reflection")
        ax.plot([centroid[0], xr[0]], [centroid[1], xr[1]], "--", color="purple", alpha=0.8)

    if "expansion" in tp:
        xe, fxe = tp["expansion"]
        ax.scatter(*xe, color="magenta", s=100, marker="P", zorder=6, label="expansion")

    if "outside_contraction" in tp:
        xoc, fxoc = tp["outside_contraction"]
        ax.scatter(*xoc, color="brown", s=90, marker="s", zorder=6, label="outside contraction")

    if "inside_contraction" in tp:
        xic, fxic = tp["inside_contraction"]
        ax.scatter(*xic, color="gray", s=90, marker="s", zorder=6, label="inside contraction")

    if annotate:
        ax.set_title(f"Iteration {step['iter']}: action = {step['action']}")
    ax.legend(loc="best", fontsize=10)
    plt.show()
```

## 7. Visualize the first few iterations in detail

These are the key figures for teaching the method. In each plot:

- **blue** is the simplex before the step,
- **green** is the simplex after the step,
- **red x** is the worst vertex,
- **black star** is the centroid of the non-worst vertices,
- colored markers show the reflection / expansion / contraction candidates.

```python
n_show = min(6, len(trace["history"]))
for k in range(n_show):
    plot_nm_step(trace["history"][k])
```

## 8. A compact textual summary of the first few steps

```python
def fmt_point(x):
    return np.array2string(np.asarray(x), precision=4, separator=", ")

for step in trace["history"][:6]:
    print("=" * 72)
    print(f"Iteration {step['iter']}")
    print("Simplex before:")
    for j, (x, fx) in enumerate(zip(step["simplex_before"], step["f_before"])):
        print(f"  vertex {j}: x = {fmt_point(x)}, f = {fx:.6f}")
    print(f"Centroid           : {fmt_point(step['centroid'])}")

    for name, (pt, val) in step["trial_points"].items():
        print(f"{name:20s}: x = {fmt_point(pt)}, f = {val:.6f}")

    print(f"Accepted action    : {step['action']}")
    print("Simplex after:")
    for j, (x, fx) in enumerate(zip(step["simplex_after"], step["f_after"])):
        print(f"  vertex {j}: x = {fmt_point(x)}, f = {fx:.6f}")
```

## 9. What the method is doing geometrically

A useful way to interpret the figures is:

- The simplex acts as a **local geometric probe** of the objective.
- The **worst** vertex is the one the method wants to replace.
- The **centroid of the better vertices** defines a direction of improvement.
- The **reflection point** tests whether moving across the centroid away from the worst point helps.
- If reflection is very good, **expansion** tries to go farther.
- If reflection is not convincing, **contraction** takes a more conservative move.
- If the simplex appears badly placed, **shrink** collapses it toward the current best point.

So Nelder-Mead is not using gradients; instead, it is using the *shape and objective ranking of the simplex* to infer a descent geometry.

## 10. Compare with SciPy's `minimize(..., method="Nelder-Mead")`

Now we also run the public SciPy optimizer from the same initial point. This is the production interface one would usually call in practice.

```python
res = minimize(
    branin,
    x0=x0,
    method="Nelder-Mead",
    options={
        "initial_simplex": initial_simplex,
        "xatol": 1e-8,
        "fatol": 1e-8,
        "maxiter": 200,
        "return_all": True,
        "disp": True,
    },
)

print(res)
```

The teaching trace above is intended for visualization. The SciPy call above is the actual library solver. Since both use the same classical Nelder-Mead logic and the same initial simplex, their qualitative behavior should match closely.

## 11. Plot the trajectory of the best point

```python
best_path = np.array([step["simplex_after"][0] for step in trace["history"]])
best_vals = np.array([step["f_after"][0] for step in trace["history"]])

fig, ax = plt.subplots(figsize=(8, 6))
contour_background(ax)
ax.plot(best_path[:, 0], best_path[:, 1], "-o", lw=2, ms=5, label="best-vertex path")
ax.scatter(*initial_simplex[0], s=120, marker="s", label="initial point")
ax.scatter(*trace["x"], s=140, marker="*", label="best after traced steps")
ax.set_title("Path of the best simplex vertex")
ax.legend()
plt.show()
```

## 12. Objective-value history

```python
plt.figure(figsize=(8, 5))
plt.plot(best_vals, "-o")
plt.xlabel("Iteration")
plt.ylabel("Best function value in simplex")
plt.title("Best-objective history over traced steps")
plt.yscale("log")
plt.show()
```

## 13. One more detailed plot: before-and-after simplex transition

```python
fig, ax = plt.subplots(figsize=(8, 6))
contour_background(ax)

for k, step in enumerate(trace["history"][:6]):
    simp = step["simplex_before"]
    cyc = np.vstack([simp, simp[0]])
    ax.plot(cyc[:, 0], cyc[:, 1], "-o", alpha=0.65, label=f"iter {k}")

ax.set_title("Evolution of the simplex over the first few iterations")
ax.legend(loc="upper right", fontsize=9)
plt.show()
```

## 14. Pedagogical takeaway

For teaching purposes, the main ideas to emphasize are:

- Nelder-Mead evolves a **simplex**, not a line-search direction.
- The simplex geometry itself carries the local information.
- Reflection is the default exploratory move.
- Expansion rewards an especially good reflection.
- Contraction is the cautious fallback.
- Shrink is the reset mechanism when the simplex is poorly aligned with local improvement.

On a 2D function like Branin-Hoo, these operations can be seen directly in the contour plots, which makes the method much easier to understand than from formulas alone.

## 15. Suggested exercises

A useful extension is to modify one of the following and rerun the notebook:

1. change the starting point;
2. change the initial simplex shape;
3. compare against coordinate search or pattern search;
4. inspect what happens near one of Branin's three global minimizers;
5. test a badly scaled objective and see how the simplex deforms.

Those changes usually make the strengths and weaknesses of Nelder-Mead very clear.
